# Juntar ficheros en uno único

Para poder realizar el estudio, vamos a juntar todos los ficheros (con datos diarios de cada año) en un único.

```python
# Librerias 
import os
import pandas as pd
```

```python
# Cargar en una lista los nombres de todos los archivos
fileList = os.listdir()

# Dejar solo los archivos  que son de tipo csv
listCsv = [file for file in fileList if file[-3:] == "csv"]

# Crear dataframe vacio
df = pd.DataFrame()

# Recorrer la lista de archivos y unirlos
for file in listCsv:
    ftemp = pd.read_csv(file, delimiter = ";")
    df = pd.concat([df, ftemp], axis=0)
```

```python
# Cabecero de los datos
df.head(5)
```

Ahora, exportamos los datos a un csv para poder trabajar con ellos.

```python
# Exportamos a csv los datos
df.to_csv('datos_total.csv', encoding='utf-8')
```
